##ImputaciÃ³n de datos

```python
import pandas as pd
import numpy as np

def imputar_promedio_entre_nans(df, max_nulos=24):
    df_resultado = df.copy()

    for columna in df.columns:
        while df_resultado[columna].isnull().any():
            df_anterior = df_resultado.copy()
            df_resultado[columna] = imputar_promedio_entre_nans_columna(df_resultado[columna], max_nulos)

            # Verificar si la imputaciÃ³n no cambia nada
            if df_resultado[columna].equals(df_anterior[columna]):
                break

    return df_resultado

def imputar_promedio_entre_nans_columna(columna, max_nulos=24):
    indices_nulos = columna.index[columna.isnull()].tolist()
    indices_promedio = []

    for i in range(len(indices_nulos)):
        if i == 0:
            inicio_bloque = indices_nulos[i]
            fin_bloque = indices_nulos[i]
        else:
            if indices_nulos[i] == indices_nulos[i-1] + 1:
                fin_bloque = indices_nulos[i]
            else:
                if fin_bloque - inicio_bloque + 1 <= max_nulos:
                    indices_promedio.append((inicio_bloque, fin_bloque))
                inicio_bloque = indices_nulos[i]
                fin_bloque = indices_nulos[i]

    if fin_bloque - inicio_bloque + 1 <= max_nulos:
        indices_promedio.append((inicio_bloque, fin_bloque))

    for inicio, fin in indices_promedio:
        indice_medio = inicio + (fin - inicio) // 2
        valor_anterior = columna.iloc[inicio - 1] if inicio > 0 else columna.iloc[fin + 1]
        valor_siguiente = columna.iloc[fin + 1] if fin + 1 < len(columna) else columna.iloc[inicio - 1]
        promedio = (valor_anterior + valor_siguiente) / 2
        columna.at[indice_medio] = round(promedio, 2)

    return columna

# imputar promedio entre NaNs en todas las columnas hasta que no haya valores nulos
df_norte_Escobedo = imputar_promedio_entre_nans(norte_Escobedo)
df_noreste2_Apodaca = imputar_promedio_entre_nans(noreste2_Apodaca)
df_noreste_SanNicolas = imputar_promedio_entre_nans(noreste_SanNicolas)
```

##ClasificaciÃ³n de los contaminantes mediante promedio mÃ³vil

```python
# Definir la funciÃ³n para calcular el promedio mÃ³vil de las Ãºltimas 12 horas
def calcular_promedio_movil(df, columna, horas):
    return df[columna].rolling(window=horas, min_periods=1).mean()

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'PM10', 12)
criterios = [0, 50, 75, 155, 235, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['PM10_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['PM10_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'PM10', 12), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['PM10_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'PM10', 12), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'PM2.5', 12)
criterios = [0, 25, 45, 79, 147, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['PM2.5_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['PM2.5_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'PM2.5', 12), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['PM2.5_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'PM2.5', 12), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'O3', 1)
criterios = [0, 0.051, 0.095, 0.135, 0.175, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['O3_1hr_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['O3_1hr_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'O3', 1), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['O3_1hr_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'O3', 1), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'O3', 8)
criterios = [0, 0.051, 0.07, 0.092, 0.114, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['O3_8hr_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['O3_8hr_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'O3', 8), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['03_8hr_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'O3', 8), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'NO2', 1)
criterios = [0, 0.107, 0.210, 0.230, 0.250, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['NO2_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['NO2_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'NO2', 1), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['NO2_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'NO2', 1), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'SO2', 24)
criterios = [0, 0.008, 0.11, 0.165, 0.22, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['SO2_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['SO2_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'SO2', 24), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['SO2_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'SO2', 24), bins=criterios, labels=etiquetas)

# Discretizar cada columna en cada DataFrame
promedio_movil = calcular_promedio_movil(df_norte_Escobedo, 'CO', 8)
criterios = [0, 8.75, 11, 13.3, 15.5, float('inf')]
etiquetas = ['Buena', 'Aceptable', 'Mala', 'Muy Mala', 'Extremadamente Mala']

# Crear nueva columna de clasificaciÃ³n para cada DataFrame
df_norte_Escobedo['CO_Clasificacion'] = pd.cut(promedio_movil, bins=criterios, labels=etiquetas)
df_noreste2_Apodaca['CO_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste2_Apodaca, 'CO', 8), bins=criterios, labels=etiquetas)
df_noreste_SanNicolas['CO_Clasificacion'] = pd.cut(calcular_promedio_movil(df_noreste_SanNicolas, 'CO', 8), bins=criterios, labels=etiquetas)
```

## CreaciÃ³n de los dataframes de San Nicolas segÃºn el viento de Apodaca y Escobedo

```python
# Agregar las 3 horas al conjunto de fechas y horas de Escobedo
# Crear un conjunto para almacenar las fechas y horas desplazadas 3 horas
fechas_escobedo = set()

# Suponiendo que 'date' es la columna que contiene las fechas en df_norte_Escobedo_filtrado
for fecha in df_norte_Escobedo_filtrado['date']:
    fecha_obj = pd.Timestamp(fecha)  # Crear un objeto Timestamp directamente
    fechas_escobedo.add(fecha_obj)  # Agregar la fecha original al conjunto
    # Agregar las 3 horas posteriores al conjunto
    for i in range(1, 4):
        nueva_fecha = fecha_obj + timedelta(hours=i)
        fechas_escobedo.add(nueva_fecha)

# Convertir el conjunto a una lista y ordenar las fechas de manera ascendente
fechas_escobedo = sorted(list(fechas_escobedo))

print(fechas_escobedo)

# Filtrar registros de San Nicolas basÃ¡ndote en el conjunto de fechas y horas de Escobedo
df_SanNico_Escobedo = df_noreste_SanNicolas_limpiado[df_noreste_SanNicolas_limpiado['date'].isin(fechas_escobedo)]

# Suponiendo que ambos DataFrames tienen una columna llamada 'date'
fechas_apodaca = set()

# Suponiendo que 'date' es la columna que contiene las fechas en df_norte_Escobedo_filtrado
for fecha in df_noreste2_Apodaca_filtrado['date']:
    fecha_obj = pd.Timestamp(fecha)  # Crear un objeto Timestamp directamente
    fechas_apodaca.add(fecha_obj)  # Agregar la fecha original al conjunto
    # Agregar las 12 horas posteriores al conjunto
    for i in range(1, 4):
        nueva_fecha = fecha_obj + timedelta(hours=i)
        fechas_apodaca.add(nueva_fecha)

# Convertir el conjunto a una lista y ordenar las fechas de manera ascendente
fechas_apodaca = sorted(list(fechas_apodaca))

# Filtrar registros de San Nicolas basÃ¡ndote en el conjunto de fechas y horas de Escobedo
df_SanNico_Apodaca = df_noreste_SanNicolas_limpiado[df_noreste_SanNicolas_limpiado['date'].isin(fechas_apodaca)]

# Suponiendo que ambos DataFrames tienen una columna llamada 'date'
df_SanNico_ApoEsco = pd.merge(df_SanNico_Apodaca, df_SanNico_Escobedo[['date']], on='date', how='inner')

# Suponiendo que ambos DataFrames tienen una columna llamada 'date'
df_SanNicolas_SinApoEsco = df_noreste_SanNicolas_limpiado.merge(
    df_SanNico_Apodaca[['date']], on='date', how='outer', indicator='indicator_apodaca'
).merge(
    df_SanNico_Escobedo[['date']], on='date', how='outer', indicator='indicator_escobedo'
)

# Filtrar las filas que no tienen coincidencias en ambos DataFrames
df_SanNicolas_SinApoEsco = df_SanNicolas_SinApoEsco[
    (df_SanNicolas_SinApoEsco['indicator_apodaca'] == 'left_only') & (df_SanNicolas_SinApoEsco['indicator_escobedo'] == 'left_only')
]

# Eliminar las columnas de indicadores
df_SanNicolas_SinApoEsco = df_SanNicolas_SinApoEsco.drop(['indicator_apodaca', 'indicator_escobedo'], axis=1)
```
